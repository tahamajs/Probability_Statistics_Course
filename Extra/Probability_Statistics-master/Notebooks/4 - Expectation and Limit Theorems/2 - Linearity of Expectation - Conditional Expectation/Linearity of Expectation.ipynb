مقدمه
                        <hr>
        </font>
        <br>
همانطور که تابحال دیده‌اید، امید ریاضی ویژگی‌های بسیاری دارد که هرکدام در جای خود کاربردهایی در اختیار ما می‌گذارند.
       در این دفترچه اما قرار است به یکی از مهم‌ترین آنها نگاه جزئی بیاندازیم.
        <br><br>
پیش از هر چیزی فکر کنید که می‌خواهید ۱۰ بار یک سکه را پرتاب کنید و به ازای هر «خط»ی که می‌آید یک آپارتمان جایزه بگیرید. برای خاطر خانه‌ها هم که شده هر چه سریعتر سکه‌ها را می‌اندازید و در نهایت اگر کمتر از ۵ خانه بگیرید زمین و زمان را لعنت می‌کنید که «من بدشانس بودم». درواقع انتظار داشتیم ۵ سکه بیاوریم تا نتوانیم بگوییم بدشانس بوده‌ایم.
قابل حدس است که در ذهنمان هربار سه‌دنگ یک سند را تصور کرده‌ایم و ۱۰ تا نصف آپارتمان را روی‌ هم گذاشته‌ایم و به میانگین/انتظار ۵ آپارتمان رسیده‌ایم.
</div>

بیان قضیه
        </font>
		<p></p>
		<hr>
حالا می‌توانید حدس بزنید که امیدریاضی «جمع چندین متغیر» چگونه محاسبه خواهد شد.
ابتدا صورت ساده‌ی قضیه را بیان و سپس آن‌را اثبات می‌کنیم.
$$
E(X+Y) = E(Y) + E(Y)
$$
اثبات:
$$E(X+Y) = \sum_{x_i, y_j}{(x_i+y_j)P_{XY}(x_i,y_j)}$$
$$= \sum_{x_i}\sum_{y_j}{x_iP_{XY}(x_i,y_j)} + \sum_{y_j}\sum_{x_i}{y_iP_{XY}(x_j,y_i)}$$
$$= \sum_{x_i}{x_iP(x_i)} + \sum_{y_i}{y_iP(y_i)} = E(X) + E(Y)$$
<br>
بطور کلی برای چندین متغیر ثابت می‌شود:
<br>
$$E(\sum{c_iX_i}) = \sum{c_iE(X_i)}$$
برای مثال 
$$E(2X+3Y-5) = 2E(X) + 3E(Y) - 5 $$
خواهد بود.
<br>
تمرین:
سعی‌کنید این اثبات را برای دو متغیر پیوسته انجام‌دهید.

متغیرهای نشان‌گر - Indicator Variables
                        <hr>
            <br>
        </font>
        یکی دیگر از کاربردهای جالب این ویژگی در مفهومی به نام متغیرهای نشانگر  مشاهده خواهد‌شد.
        <br><br>
        متغیرهای نشانگر متغیرهایی هستند که به یک واقعه نسبت داده می‌شوند. اگر این واقعه اتفاق بیفتد مقدار یک و در غیر این صورت مقدار صفر را خواهند گرفت.
        این تغییر دیدگاه کوچک، ما را در حل مسائل کمک بسیاری خواهد کرد.
        <br>
        $$
        X = \left\{
            \begin{array}{l}
            1 & \mbox{happens} \\
            %1 & 1 \lt a \le 2 \\
            0 & \mbox{otherwise}
            \end{array}
            \right.
         $$
         <br>
        سعی می‌کنیم در قالب مثال آن را باهم بفهمیم:
    <br>
فرض کنید ۵ سکه برای پرتاب‌کردن در اختیار داریم. میخواهیم بدانیم بطور میانگین بعد از ۵ بار پرتاب چند بار دو بار پشت سر هم «خط» (T) خواهد آمد.
<br>
دقت کنید که اگر در پرتاب‌ها نتیجه‌ی TTTHT بیاید، ۲ جفت شمارش خواهیم‌کرد.
<br><br>
<font color="06488a" size=4>
<p style="text-align: center;">
*اگر برای بار اول این جزوه را می‌خوانید حتماً سعی‌کنید روی این مثال چند دقیقه وقت بگذارید و به راه‌حل‌های مختلف آن فکرکنید.*
<br>
    </p>
</font>
<br>
مسأله را از این دید نگاه می‌کنیم که اتفاقاتی قرار است پشت سر هم رخ بدهد و هر بار که این اتفاق رخ داد یک امتیاز دریافت‌کنیم.
بدین ترتیب فکر می‌کنیم که تعدادی جفت سکه در اختیار داریم و قرار است آن‌هارا چندین بار پشت سر هم پرتاب کنیم.
با داشتن ۵ سکه، ۴ جفت سکه برای حالت‌گیری در اختیار خواهیم داشت (توجه به شکل).
در کل اگر هر جفت را بصورت $x_{ij}$ در نظر بگیریم جواب مسئله برابر خواهد بود با:

$$
E(X_{12}+X_{23}+X_{34}+X_{45}) = E(X_{12}) + E(X_{23}) + E(X_{34}) + E(X_{45})$$$$= 4{E(X_{12})} =
4(1.P(happened) + 0.P(\mbox{not happened})) = \\
4(1.(\frac{1}{2}.\frac{1}{2})+ 0) = 1$$
<br>

<div style="float:center">
$E(X_{ij})$ ها مقادیر یکسانی دارند و آن‌هارا یکسان و برابر با $E(X_{12})$ گرفتیم.
</div>

<img src="figs/coins.png" style="float:center; width:50%"/>
شاید برای شما گسستن این رویداد‌های پشت سر هم و وابسته به جمع ۴ متغیر جدا از هم کار درستی به نظر نیاید اما توجه کنید 
در اینجا
<font color="#287340">
ویژگی خطی بودن امیدریاضی به کمک ما آمده که هیچ ارتباطی به وابستگی یا استقلال متغیرهای ورودی‌اش ندارد</font>!
</font>
</div>

مثال دیگر:
        </font>
		<p></p>
        در شروع یک مهمانی، N مرد کلاه‌هایشان را تحویل می‌دهند. مسول نگهداری کلاه‌ها دچار مسمومیت می‌شود و کلاه‌ها با هم قاطی می‌شوند و در نتیجه در انتهای مهمانی، هر فرد، یک کلاه تصادفی را دریافت می‌کند.
در واقع هر کس به احتمال 
        $\frac{1}{N}$
        کلاه خودش را دریافت می‌کند.
        <br>
        تعداد افرادی را که انتظار داریم کلاه خودشان را دریافت کرده باشند چقدر است؟ 
       <br>
<font size=6>
پاسخ:
        </font>
بدون استفاده از متغیر‌های نشان‌گر، به سختی قادر به حل این مساله خواهیم بود!اما با استفاده از متغیر‌های نشان‌گر، این مساله به سادگی حل می‌شود. 
        $I_i$ 
       را به این شکل تعریف می کنیم:
       $$
      I_i = \left\{
            \begin{array}{l}
            1 & \mbox{the i-th man gets his own hat} \\
            %1 & 1 \lt a \le 2 \\
            0 & \mbox{o.w.}
            \end{array}
            \right.
       $$
        <br>
       حال اگر متغیر تصادفی X برابر با تعداد افرادی باشد که کلاه خودشان را دریافت می‌کنند، خواهیم داشت:
        $$
        X = \sum_{i=1}^{N}I_i
        $$
        خواسته‌ی مساله از ما، امید ریاضی این متغیر تصادفی می‌باشد، پس خواهیم داشت:
        $$\mathop{\mathbb{E}}[X] = \mathop{\mathbb{E}}[\sum_{i=1}^{N}I_i]$$
        که با استفاده از خطی بودن امید ریاضی تبدیل می‌شود به:
        $$\mathop{\mathbb{E}}[X]= \sum_{i=1}^{N} \mathop{\mathbb{E}}[I_i]$$
        $$\mathop{\mathbb{E}}[X] = \sum_{i=1}^{N} (\frac{1}{N}) = 1$$
        
</font>
</div>

<div id="sec_ind_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=##f5161a size=6.5>
مثال چندبخشی و مهم!
        </font>
		<p></p>
        فرض‌کنید گروهی ۵۰ نفره داریم که روز‌های تولد مستقل از هم دارند. میخواهیم: <br>
        الف)امید ریاضی افرادی که تاریخ تولد
        یکسانی
        دارند و 
       <br>
ب) امید ریاضی تعداد روز‌هایی از سال که حداقل ۲ نفر در آن متولد‌شده‌اند
<br>
را محاسبه‌کنیم.
<br><br>
        الف)
        <br>
        به مثال قبل فکر کنید. میخواهیم مسئله را به فضای «اتفاق افتادن یا نیفتادن یک رویداد برای چندین شئ مشابه» ببریم.
     در اینجا، اتفاقات برابر خواهند بود با اینکه آیا هر فرد با حداقل یک فرد دیگر از ۴۹ نفر باقی‌مانده «هم‌تولد» باشد یا خیر. اگر کس دیگری با آنها در روز تولد اشتراک داشت شخص مورد سوال را «دایی» می‌نامیم.
        <br>
        بنابراین ۵۰ نفر را به صف می‌ایستانیم(!) و از هرکدامشان وضعیت «دایی بودن» را می‌پرسیم. اگر فرض کنیم احتمال دایی بودن به ما اعلام‌شده‌باشد، میانگین در بین همه برابر خواهد بود با ۵۰ برابر احتمال دایی بودن.
        چون این احتمال برای همه‌ی افراد یکسان است آن را $P(x)$ می‌نامیم و محاسبه‌اش می‌کنیم.$$
        f_{X}(x) = \left\{
            \begin{array}{l}
            1 & \mbox{has at least one birth-mate} \\
            %1 & 1 \lt a \le 2 \\
            0 & \mbox{o.w.}
            \end{array}
            \right.
         $$
       <br>
        $$
        P(X) = 1 - P(X)'$$
        $$
        P(X) = 1 - (P(Y))^{49}$$ 
        <font size="4">
*        $Y$ را برابر با متفاوت بودن روز تولد فرد بخصوصی از فرد X در نظر بگیرید.
        </font>
        $$
        P(X) = 1 - (364/365)^{49}$$$$
        = 1 - 0.87 = 0.13
        $$:بدین ترتیب
        <br>
        $$
       E(X) = E(X_1 + X_2 + ... + X_{50}) = 50(X(u_1))$$$$ = 50 (1\times{P(X_1)} + 0\times{P(X_1)'}) = 50*0.13 = 6.5$$
<br>
<img src ="figs/queue.jpg" style="float:center; border-radius:2%;">
<br>
ب)
<br>
حال کافی‌ست افراد مورد سوال را روز‌های سال قرار دهیم! از روز اول نوروز سوال می‌کنیم که آیا ۲ نفر یا بیشتر از آن ۵۰ نفر در او متولد‌شده‌اند یا خیر. بهمین ترتیب تا روز ملی‌شدن صنعت نفت پیش می‌رویم (۳۶۵ روز):
<br>
$$
E(X) = E(X_1+X_2+...X_{365}) $$$$
E(X) = \sum_{i = 1}^{365}X_i = 365\times{P(X)}$$$$
$$
<br>
$P(X)$ برابر است با احتمال اینکه بیش از یک نفر در یک روز خاص از سال متولد شده‌باشند.
<br>
$$P(X) = 1 - P(X)' = 1 - (P(\mbox{zero person born}) + P(\mbox{one person born}))$$
$$P(X) = 1 - ((\frac{364}{365})^{50} + {50 \choose 1}\times(\frac{364}{365})^{49}\times\frac{1}{365})$$
$$P(X) \approx 1 - 0.991 = 0.009$$
<br>
$$E(X) = 365*0.009 = 3.3$$
