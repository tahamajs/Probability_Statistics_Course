امید ریاضی شرطی
        </font>
		<p></p>
		<hr>
در مبحث احتمال شرطی یادگرفتیم که می‌توان احتمال رخ‌دادن یک رویداد را در فضاهای محدود‌تری محاسبه‌کرد. فضاهای شرطی به ما این اجازه را می‌دهند که با محاسبه‌کردن حالت‌های خاص‌تری از رویداد‌های کلی محاسبات برای ما ساده‌تر شود.
در حالت کلی تر توزیع احتمال شرطی دو متغیر مختلف
را به صورت‌های زیر در هر دو حالت گسسته و پیوسته تعریف کردیم:
        <br>
$$
p_{X|Y}(x|y) = P\{X=x|Y=y\} = \frac{P(x,y)}{P_Y(y)} \quad (P_Y(y) \neq 0)
$$
        <br>
$$
f_{X|Y}(x|y) = \frac{f(x,y)}{f_Y(y)} \quad (f_Y(y) \neq 0)
$$
حال به کمک مفاهیم بالا می‌توان امید ریاضی شرطی را تعریف کرد.
اما پیش از آن به مثال زیر توجه کنید:
<br/>
از مثال‌های قبلی جزوه می‌دانیم که امید ریاضی متغیر تصادفی خروجی یک تاس متقارن برابر با $3.5$ است.
حال فرض کنید که ما با علم پیشین خود میدانیم که خروجی تاس عددی زوج بوده است. در این حالت آیا باز هم امید ریاضی همان عدد سابق خواهد بود؟ درست است به علم اضافی‌مان اهمیتی ندهیم؟ درست است که باز هم اعداد فرد
را دخیل کنیم؟
پاسخ منفی است.
با علم به این‌که خروجی زوج بوده است و تاس هم متقارن، به نظر می‌رسد که باید احتمال برابری برای همه خروجی‌های زوج در نظر بگیریم و سپس امید ریاضی را محاسبه کنیم.
$$
\mathop{\mathbb{E}}[X|A] = \frac{1}{3} \times 2 + \frac{1}{3} \times 4 + \frac{1}{3} \times 6 = 4
$$
که در آن
$A$
رویداد
«زوج آمدن خروجی»
است.
<br/>
        <br><br/>
حال امید ریاضی شرطی را به صورت زیر تعریف می‌کنیم:
$$
\mathop{\mathbb{E}}[X|A] = \sum_{x}{xp(X=x|A)}
$$
طبق انتظار، امید ریاضی متغیر
$X$
مشروط بر متغیر
$Y$
در دو حالت گسسته و پیوسته به شکل‌های زیر تعریف می‌شود:
<br/>
<br/>
$$
\mathop{\mathbb{E}}[X|Y=y] =   
\begin{cases}
       \sum_{x}{xP(X=x|Y=y)} = \sum_{x}{xP_{X|Y}(x|y)} &\quad \text{Discrete Case}\\
       \ & \ \\
       \int_{-\infty}^{\infty}{xf_{X|Y}(x|y)}dx &\quad \text{Continuous Case}\\ 
\end{cases}
$$
<br/>
<div style="text-align: center;">
$
P_Y(y) > 0
$
در حالت گسسته
<br>
$
f_Y(y) > 0
$
در حالت پیوسته
        </div>
</font>
</div>

<p></p>
<br />
<div id="sec_conditional_expectation_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
فرض کنید که تابع چگالی احتمال مشترک
$X$
و
$Y$
به شرح زیر است:
$$
f(x,y) = \frac{e^{-\frac{x}{y}}e^{-y}}{y} \quad 0 < x < \infty, 0 < y < \infty
$$
می‌خواهیم مقدار
$\mathop{\mathbb{E}}[X|Y=y]$
را محاسبه کنیم. 
        <br/><br/>
$$
\begin{aligned}
            f_{X|Y}(x|y) &= \frac{f(x,y)}{f_Y(y)}
            = \frac{f(x,y)}{\int_{-\infty}^{\infty}f(x,y)dx} 
            = \frac{\frac{e^{-\frac{x}{y}}e^{-y}}{y}}{{\int_{0}^{\infty}\frac{e^{-\frac{x}{y}}e^{-y}}{y}dx}} 
            = \frac{1}{y}e^{-\frac{x}{y}}
\end{aligned}
$$
        
$$
\Rightarrow \mathop{\mathbb{E}}[X|Y=y] = \int_{0}^{\infty}\frac{x}{y}e^{-\frac{x}{y}}dx = y
$$
</font>
</div>

<div id="sec_conditional_prop" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
ویژگی‌های امید ریاضی شرطی
        </font>
		<p></p>
		<hr>
امید ریاضی شرطی ویژگی‌های امید ریاضی معمولی (غیرشرطی) را دارد. به عبارت دیگر امید ریاضی شرطی 
$
\mathop{\mathbb{E}}[X|Y=y]
$
همان امید ریاضی معمولی است که فضای نمونه آن
به
زیرفضایی که رویدادهای
آن
$Y=y$
هستند،
کاهش پیدا کرده است.
<br><br> 
<b>
    ویژگی ۱:
        </b>
        تابعی از متغیر تصادفی مثل قبل محاسبه می‌شود!
$$
\mathop{\mathbb{E}}[g(X)|Y=y] = \begin{cases}
    \sum_{x}g(x)P_{X|Y}(x|y) &\quad \text{ Discrete Case}\\
    \int_{-\infty}^{\infty}g(x)f_{X|Y}(x|y)dx &\quad \text{ Continuous Case}
\end{cases}
$$
<br><br> 
        <b>
ویژگی ۲:
        </b>
        خطی بودن همچنان برقرار است!
$$
\mathop{\mathbb{E}}[(\sum_{i=1}^{n}X_i)|Y=y] = \sum_{i=1}^{n}\mathop{\mathbb{E}}[X_i|Y=y]
$$
<br/>        
حال عبارت
و نمادگذاری
$\mathop{\mathbb{E}}[X|Y]$
را به عنوان تابعی از متغیر تصادفی
$Y$
<strong>تعریف</strong> می‌کنیم که مقدار آن در هر نقطه مانند
$Y=y$
برابر
$\mathop{\mathbb{E}}[X|Y=y]$
باشد. به عبارت دیگر در این نمادگذاری با حذف مقدار
$y$
از امید ریاضی شرطی (بر خلاف امید ریاضی که یک مقدار دقیق و مشخص است)
به یک متغیر تصادفی بر حسب
$Y$
رسیدیم.
<br/>
حال چون
$\mathop{\mathbb{E}}[X|Y]$
متغیر تصادفی است،
بنابراین امید ریاضی هم روی آن تعریف می‌شود.
ویژگی بسیار مهم زیر اهمیت امید ریاضی گرفتن از
$\mathop{\mathbb{E}}[X|Y]$
را مشخص می‌کند:
<br><br>
        <font color="red">
            <b>
ویژگی ۳:
            </b>
        </font>
قانون کل امید ریاضی (law of total expectation)
$$
\mathop{\mathbb{E}}[X] = \mathop{\mathbb{E}}[\mathop{\mathbb{E}}[X|Y]]
$$
<br><br/>
به عبارت دیگر در دو حالت گسسته و پیوسته داریم:
$$
\mathop{\mathbb{E}}[X] = \sum_{y}\mathop{\mathbb{E}}[X|Y=y]P\{Y=y\}
$$
$$
\mathop{\mathbb{E}}[X] = \int_{-\infty}^{\infty}\mathop{\mathbb{E}}[X|Y=y]f_Y(y)dy
$$
<font color="7a1f7a">
در واقع برای محاسبه امید ریاضی
$X$
می‌توان آن را به شرط مقدارهای مختلف
$Y$
حساب کرد. حال در نهایت این مقادیر باید در احتمال وقوع هر
$Y$
نیز ضرب شوند و سپس با هم جمع شوند.
        </font>
(این ویژگی مهم کمی در نگاه اول عجیب است. لطفاً دوباره آن را بخوانید.)
<br/>
اثبات در حالت گسسته:
$$
\begin{aligned}
\sum_{y}\mathop{\mathbb{E}}[X|Y=y]p\{Y=y\} &= \sum_{y}\sum_{x}xP\{X=x|Y=y\}P\{Y=y\} \\
&= \sum_{y}\sum_{x}x\frac{P\{X=x, Y=y\}}{P\{Y=y\}}P\{Y=y\} \\
&= \sum_{y}\sum_{x}xP\{X=x,Y=y\}\\
&= \sum_{x}x\sum_{y}P\{X=x,Y=y\}\\
&= \sum_{x}xP\{X=x\}\\
&= \mathop{\mathbb{E}}[X]
\end{aligned}
$$
<br>
تمرین: سعی کنید عبارت بالا را برای حالت پیوسته هم ثابت کنید. 
</font>
</div>

<div id="sec_conditional_prop_example2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال 
۱
        </font>
		<hr>
 یک معدنچی در معدنی که ۳ درب دارد گیر افتاده است. درب اول او را به تونلی هدایت می‌کند که پس از طی مسیری ۲ ساعته، او را  به خارج از معدن می‌رساند. درب دوم او را به تونلی می‌برد که پس از ۳ ساعت وی به همان نقطه اول باز می‌گردد. درب سوم هم مانند درب دوم است ولی ۵ ساعت طول می‌کشد تا معدنچی به نقطه اولش بازگردد. با فرض این‌که به دلیل تاریکی معدنچی نمی‌تواند درب‌ها را از هم تشخیص بدهد و هر بار با احتمال مساوی یکی از درب‌ها را انتخاب کند، امید ریاضی مدت زمانی که طول می‌کشد تا او از معدن خارج شود چند است؟
<img src="figs/mine.png" style="float:center; width:30%"/>
<br/>
<br/>
مدت زمان مد نظر را
با متغیر تصادفی
$X$
نمایش می‌دهیم.
سپس متغیر
$Y$
را برابر دربی که معدنچی به تصادف انتخاب می‌کند در نظر می‌گیریم.
در نتیجه در زمانی که معدنچی شروع به حرکت می‌کند، داریم:
<br><br>
$$
\mathop{\mathbb{E}}[X] = \mathop{\mathbb{E}}[X|Y=1].P[Y=1] + \mathop{\mathbb{E}}[X|Y=2].P[Y=2] + \mathop{\mathbb{E}}[X|Y=3].P[Y=3] = \frac{1}{3}[\mathop{\mathbb{E}}[X|Y=1] + \mathop{\mathbb{E}}[X|Y=2] + \mathop{\mathbb{E}}[X|Y=3]] = \frac{1}{3}[(2) + (\mathop{\mathbb{E}}[X + 3]) + (\mathop{\mathbb{E}}[X + 5])]
$$
$$
\to \mathop{\mathbb{E}}[X] = \frac{1}{3}[(2) + (\mathop{\mathbb{E}}[X] + 3) + (\mathop{\mathbb{E}}[X] + 5)]
$$
$$
\to \mathop{\mathbb{E}}[X] = 10
$$
<br>
</font>
</div>

<div id="sec_conditional_prop_example2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال 
۲
        </font>
		<hr>
  مساله قدم زدن تصادفی را در نظر بگیرید. فردی روی یک عدد طبیعی تصادفی بین $0$ تا $N$ قرار دارد و در هرگام با احتمال $\frac{1}{2}$ یک قدم به سمت راست یا چپ بر میدارد و زمانی که به یکی از دو انتهای بازه($0$ یا $N$) برسد کار خود را تمام می‌کند. می‌خواهیم متوسط زمانی که طول می‌کشد تا فرد به یکی از دو انتها برسد را حساب کنیم.
<br>
پاسخ: $T$ را به عنوان زمان رسیدن به یکی از دو انتها درنظر بگیرید. مشخصا این یک کمیت تصادفی است و بنابراین امید ریاضی روی آن تعریف می‌شود. فرض میکنیم فرد در ابتدا روی عدد $i$باشد و زمان رسیدن به یکی از دو انتها را با $T_i$ نشان می‌دهیم. اگر این متغیر را روی جهت حرکت(به سمت راست یا چپ) شرطی کنیم، طبق ویژگی 
        $$
\mathop{\mathbb{E}}[X] = \mathop{\mathbb{E}}[\mathop{\mathbb{E}}[X|Y]]
$$ خواهیم داشت:<br>
        $$
        \mathop{\mathbb{E}}[T_i] = \mathop{\mathbb{E}}[\mathop{\mathbb{E}}[T_i|\text{direction}]] = P[\text{left}]\mathop{\mathbb{E}}[T_i|\text{left}]] + P[\text{right}] \mathop{\mathbb{E}}[T_i|\text{right}]
        $$
<br>
وقتی به سمت راست یا چپ حرکت میکنیم در واقع یک قدم برداشته‌ایم و در یک جایگاه جدید قرار داریم. پس متوسط رسیدن به انتها از این جایگاه جدید هرچه باشد، با یک قدمی که برداشته‌ایم جمع خواهد یعنی 
$$
\mathop{\mathbb{E}}[T_i|\text{left}] = \mathop{\mathbb{E}}[T_{i-1}] + 1, \quad
\mathop{\mathbb{E}}[T_i|\text{right}] = \mathop{\mathbb{E}}[T_{i+1}] + 1
$$       
<br>
همچنین احتمال حرکت به سمت راست یا چپ برابر $\frac{1}{2}$ است. از این پس $\mathop{\mathbb{E}}[T_i] = a_i$ قرار می‌دهیم.
$$
a_i = \frac{1}{2} (a_{i-1} + 1) + \frac{1}{2} (a_{i+1} + 1), \quad a_0 = a_N = 0
$$
<br>
رابطه بازگشتی بالا را حل کنید و امید ریاضی زمان رسیدن به یکی از دو انتها با شروع از نقطه $i$ را به دست بیاورید.
</font>
</div>

<div id="thi_conditional_prop_example3" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال 
۳
        </font>
		<hr>
        واقعه‌ای که شرطمان روی آن مطرح می‌شود، می‌تواند مربوط به همان متغیر تصادفی‌ای باشد که می‌خواهیم امید ریاضی آن را بدست آوریم.
        <br>
        برای مثال فرض کنید 
        $X \sim N(0, \sigma^2)$
        باشد، حال می‌خواهیم
        $Var(X|X>0)$
        را بدست آوریم. برای این‌کار خواهیم داشت:
        $$
        f_X(x) = \frac{1}{\sqrt{2\pi}\sigma} e^{-\frac{x^2}{2\sigma^2}}
        $$
        $$
        f_{X|X>0}(x|x>0) = \left\{
            \begin{array}{l}
            \frac{f_X(x)}{1 - F_X(0)} = 2f_X(x) & \mbox{x > 0} \\
            %1 & 1 \lt a \le 2 \\
            0 & \mbox{o.w.}
            \end{array}
            \right.
        $$
        و همچنین داریم:
        $$
        Var(X|X>0) = \mathop{\mathbb{E}}[X^2|X>0] - \mathop{\mathbb{E}}[X|X>0]^2
        $$
        حال هر کدام از امید ریاضی‌ها را حساب می‌کنیم تا به نتیجه‌ی نهایی برسیم:
        $$
        \mathop{\mathbb{E}}[X^2|X>0] = \int_{0}^{\infty}x^2 f_X(x)dx = \int_{0}^{\infty} x^2 \frac{2}{\sqrt{2\pi}\sigma} e^{-\frac{x^2}{2\sigma^2}} dx
        $$
        $$
        \Rightarrow\mathop{\mathbb{E}}[X^2|X>0] = \sigma ^ 2
        $$
        $$
        \mathop{\mathbb{E}}[X|X>0] = \int_{0}^{\infty}x f_X(x)dx = \int_{0}^{\infty} x \frac{2}{\sqrt{2\pi}\sigma} e^{-\frac{x^2}{2\sigma^2}} dx
        $$
        $$
        \Rightarrow\mathop{\mathbb{E}}[X|X>0] = \sqrt{\frac{2}{\pi}} \sigma
        $$
        $$
        \Rightarrow\mathop{\mathbb{E}}[X|X>0]^2 = \frac{2 \sigma ^ 2}{\pi}
        $$
        پس واریانس به این شکل بدست می‌آید:
        $$
        Var(X|X>0) = (1 - \frac{2}{\pi})\sigma ^ 2
        $$
<br>
</font>
</div>
