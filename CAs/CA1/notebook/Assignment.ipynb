```python
import  numpy as np
def sample_binomial_distribution(m, n, p):
    samples = np.random.choice([0, 1], size=(m, n), p=[1-p, p])
    sums = np.sum(samples, axis=1)
    return sums


```

in this function we have m number of samples that has binoial destribution with n number of trials and p probability of success
becuse we have n number of bernouli destribution and we want to sum them up we can use binomial destribution for this problem
out put of this function is m number of samples that has binomial destribution with n number of trials and p probability of success
we have n number of bernouli destribution and we want to sum them up we can use binomial destribution for this problem .

```python
n = 500 

#for that we need 100 of p values for several p values 
#for generating P values
p_values = np.linspace(0,1,101)

# for generating 10000 samples of binomial destribution
m = 5000  


##  this are theoretical values for binomial distribution for all of p values in numpy array that named theoretical_means and theoretical_variances
theoretical_means = n * p_values  # theoretical mean for all of p values 
theoretical_variances = n * p_values * (1 - p_values)  #theoretical variance for all of p values 
```

this is example of how linespace function works
for example we want 100 numbers between 0 and 1 like part 2 of question 1
we put it 101 because that we want 0 and 1 too

```python
p_values = np.linspace(0,1,101)
print(p_values)
```

```python
import matplotlib.pyplot as plt

sample_means = np.zeros(len(p_values))
sample_variances = np.zeros(len(p_values))

for i, p in enumerate(p_values):
    # for all of 100 p values we generate 5000 samples of binomial destribution and get that samples mean and variance and put them in sample_means and sample_variances arrays
    samples = sample_binomial_distribution(m, n, p)
    sample_means[i] = np.mean(samples)
    sample_variances[i] = np.var(samples)

# plot the results
plt.figure(figsize=(20, 20))
plt.plot(p_values, theoretical_means, label="theoretical mean")
plt.plot(p_values, sample_means, label="practical mean")
plt.plot(p_values, theoretical_variances, label="theoretical variance ")
plt.plot(p_values, sample_variances, label="practical variance ")
plt.xlabel("P values")
plt.ylabel("samples")
plt.legend()
plt.title("theoretical and practical  values for binomial distribution")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# fist we generate 10000 samples of binomial distribution with n = 250 and p = 0.08
n = 250
p = 0.008

# we can use binomial function in numpy for generating binomial distribution
binomial_samples = np.random.binomial(n, p, size=10000)

# plot the results
plt.hist(binomial_samples, bins=100, density=True, alpha=1, color='g')
plt.title('Binomial Distribution (n=250, p=0.008)')
plt.xlabel('Number of Occurrences')
plt.ylabel('Probability Density')
plt.show()
```

```python
from scipy.stats import poisson


#poisson distribution with lambda = 2
poisson_lambda = 2

poisson_samples = poisson.rvs(mu=poisson_lambda, size=10000)

plt.hist(poisson_samples, bins=100, density=True, alpha=1, color='g')
plt.title('Poisson Distribution (lambda=20)')
plt.xlabel('Number of Occurrences')
plt.ylabel('Probability Density')
plt.show()
```

```python
from math import sqrt
from scipy.stats import norm

normal_mean = 2
normal_std = sqrt(1.984)

x = np.linspace(normal_mean - 3*normal_std, normal_mean + 3*normal_std, 10000)
normal_samples = norm.pdf(x, normal_mean, normal_std)

plt.plot(x, normal_samples, 'b-', lw=2)
plt.title('Normal Distribution (mean=3182.59, std=6.47)')
plt.xlabel('Number of Occurrences')
plt.ylabel('Probability Density')
plt.show()
```
