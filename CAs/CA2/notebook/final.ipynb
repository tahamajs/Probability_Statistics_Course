and we need to show that this variable is in poisson distribution.

```python
from scipy.stats import poisson

X = tarbiat_data['BRT']
Y = tarbiat_data['metro']
Z = X + Y

lambda_x = tarbiat_data['metro'].mean()
lambda_y = tarbiat_data['BRT'].mean()

lambda_z = lambda_x + lambda_y

z_samples = poisson.rvs(mu=lambda_z, size=len(Z))

print('lambda for Z is : ', Z.mean())
print('variance for Z is : ', Z.var())

print('lambda for X is : ', X.mean())
print('variance for X is : ', X.var())

print('lambda for Y is : ', Y.mean())
print('variance for Y is : ', Y.var())

print('lambda for X + Y is : ', (X + Y).mean())
print('variance for X + Y is : ', (X + Y).var())

print('we can see that lambda for X + Y is equal to lambda for Z')
```

```python

import  random

def collector_simulation(n, k):
    total_collected = 0

    for _ in range(k):
        leaflets = [0] * n

        while min(leaflets) == 0:
            selected_leaflet = random.randint(0, n - 1)
            leaflets[selected_leaflet] += 1
        total_collected += sum(leaflets)
        
        # print(leaflets)
    return total_collected / k


print('sample data : ', collector_simulation(10, 10000))
```
